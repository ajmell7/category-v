```python
import hurdat_data_manipulator
import glm_file_downloader

from datetime import datetime
```

```python
hurdat_df = hurdat_data_manipulator.HurdatDataManipulator()
glm_downloader = glm_file_downloader.GLMFileDownloader()
```

```python
df_atl = hurdat_df.create_dataframe("atl")
df_pac = hurdat_df.create_dataframe("pac")
```

```python
hurricane_ian_df = df_atl[df_atl.name == "IAN"].reset_index()
hurricane_ian_df["day_of_year"] = hurricane_ian_df["date"].dt.dayofyear
#hurricane_ian_df
```

```python
## an outer for loop could be added here to go through all hurricanes 

ian_data_paths = []
min_day = hurricane_ian_df.day_of_year.min()
max_day = hurricane_ian_df.day_of_year.max()
min_year = hurricane_ian_df.year.min()
max_year = hurricane_ian_df.year.max()
years = range(min_year, max_year + 1)
days = range(min_day, max_day + 1)
hours = range(0, 24)


for year in years:
    for day_of_year in days:
        for hour in hours:
            ian_data_paths.append(f"GLM-L2-LCFA/{year}/{day_of_year:03d}/{hour:02d}/")

#ian_data_paths
```

```python
#### Edit add check to determine which bucket name to download to
bucket_name = "gcp-public-data-goes-16"
destination_dir = "glm_files"
ian_data = None
for prefix in ian_data_paths:
    ian_data = glm_downloader.download_to_memory(bucket_name, prefix)
```
