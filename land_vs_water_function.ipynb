```python
from global_land_mask import globe
import numpy as np
import pandas as pd
```

```python
best_track_df = pd.read_csv(f'data/storms/IAN_2022/hurricane/besttrack.csv', parse_dates=['Timestamp'])
```

```python
land_values = []
for row in best_track_df.itertuples():
    lat = row.Latitude
    lon = row.Longitude
    is_land = globe.is_land(lat, lon) 
    land_values.append("Y" if is_land else "N")

best_track_df['Land'] = land_values
```
