```python
import xarray as xr
import os
import numpy as np
import pandas as pd
import ipywidgets as widgets
from datetime import datetime, timedelta

from glmtools.io.glm import GLMDataset
from glmtools.test.common import get_sample_data_path
from google.cloud import storage
from google.auth.credentials import AnonymousCredentials
```

```python
def download_blob(bucket_name, source_blob_name, destination_file_name):
    storage_client = storage.Client(credentials=AnonymousCredentials(), project="public-data")
    bucket = storage_client.bucket(bucket_name)
    blob = bucket.blob(source_blob_name)

    blob.download_to_filename(destination_file_name)
```

```python
bucket_name = "gcp-public-data-goes-18"
source_blob_name = "GLM-L2-LCFA/2025/001/00/OR_GLM-L2-LCFA_G18_s20250010000000_e20250010000200_c20250010000219.nc"
destination_file_name = "test_file.nc"
download_blob(bucket_name, source_blob_name, destination_file_name)
```
